# Correlation Analysis

```python
numerical_columns_customers = customers.select_dtypes(include=['int64', 'float64']).columns
print(numerical_columns_customers)
if len(numerical_columns_customers) > 0:
    correlation_matrix_customers = customers[numerical_columns_customers].corr()
    
    # Plot the heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix_customers, annot=True, cmap='RdYlGn', fmt='.2f')
    plt.title('Correlation Matrix for Customers Dataset')
    plt.show()
else:
    print("No numerical columns available for correlation analysis in Customers dataset.")
```

```python
numerical_columns_products = products.select_dtypes(include=['int64', 'float64']).columns
print("Numerical columns in Products dataset:", numerical_columns_products)

# Calculate the correlation matrix if there are numerical columns
if len(numerical_columns_products) > 0:
    correlation_matrix_products = products[numerical_columns_products].corr()
    
    # Plot the heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix_products, annot=True, cmap='RdYlGn', fmt='.2f')
    plt.title('Correlation Matrix for Products Dataset')
    plt.show()
else:
    print("No numerical columns available for correlation analysis in Products dataset.")
```

```python
numerical_columns_transactions = transactions.select_dtypes(include=['int64', 'float64']).columns
print("Numerical columns in Transactions dataset:", numerical_columns_transactions)

# Calculate the correlation matrix if there are numerical columns
if len(numerical_columns_transactions) > 0:
    correlation_matrix_transactions = transactions[numerical_columns_transactions].corr()
    
    # Plot the heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix_transactions, annot=True, cmap='RdYlGn', fmt='.2f')
    plt.title('Correlation Matrix for Transactions Dataset')
    plt.show()
else:
    print("No numerical columns available for correlation analysis in Transactions dataset.")
```

# Time Series Analysis

```python
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'])

# Resample to monthly total value
monthly_total = transactions.resample('M', on='TransactionDate').sum()['TotalValue']

# Time series plot
plt.figure(figsize=(12, 6))
monthly_total.plot()
plt.title('Monthly Total Value of Transactions')
plt.xlabel('Date')
plt.ylabel('Total Value (USD)')
plt.grid()
plt.show()
```

```python
moving_average = monthly_total.rolling(window=3).mean()

# Plot monthly total transaction values with moving average
plt.figure(figsize=(12, 6))
plt.plot(monthly_total, label='Monthly Total', color='blue')
plt.plot(moving_average, label='3-Month Moving Average', color='orange')
plt.title('Monthly Total Transaction Values with Moving Average')
plt.xlabel('Date')
plt.ylabel('Total Value (USD)')
plt.legend()
plt.grid()
plt.show()
```

# Data Cleaning

```python
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'])

# Check data types
print(customers.dtypes)

# Outlier Detection (if applicable)
# Example: Box plot for product prices
plt.figure(figsize=(10, 6))
sns.boxplot(x=products['Price'])
plt.title('Box Plot of Product Prices')
plt.xlabel('Price (USD)')
plt.show()

# Feature Engineering
# Example: Create a feature for total spending per customer
total_spending = transactions.groupby('CustomerID')['TotalValue'].sum().reset_index()
total_spending.rename(columns={'TotalValue': 'TotalSpending'}, inplace=True)
customers = customers.merge(total_spending, on='CustomerID', how='left')

# Display the updated customers DataFrame
print(customers.head())
```
